```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#We found the dataframe we need for our initial analysis
df=pd.read_csv("sip.csv")
```

```python
#Lets organize the columns, renaming them and dropping the ones we dont need.
df.rename(columns={'geo': 'COUNTRY', 'indic_de': 'INDEX','TIME_PERIOD': 'YEAR', 'OBS_VALUE': 'VALUE'}, inplace=True)
df=df.drop(["DATAFLOW","LAST UPDATE","OBS_FLAG","freq"],axis=1)
```

```python
#Disregard the countries we are not interested in:
df=df[~df['COUNTRY'].str.contains("TR")]
df=df[~df['COUNTRY'].str.contains("XK")]
df=df[~df['COUNTRY'].str.contains("UKC-L")]
df=df[~df['COUNTRY'].str.contains("UKN")]
df=df[~df['COUNTRY'].str.contains("AD")]
df=df[~df["COUNTRY"].str.contains("BY")]
df=df[~df["COUNTRY"].str.contains("GE")]

df
```

```python
#We have found another df with info about the GDP evolution for each country.
pib=pd.read_csv("sip2.csv")
pib
```

```python
#We wil create a function to avoid doing all the basic cleaning and formatting everytime we need to use a table from Eurostat
def clean_eurostat(df):
    df.rename(columns={'geo': 'COUNTRY', 'indic_de': 'INDEX','TIME_PERIOD': 'YEAR', 'OBS_VALUE': 'VALUE',"sex":"SEX"}, inplace=True)
    df=df.drop(["DATAFLOW","LAST UPDATE","OBS_FLAG","freq"],axis=1)
    df=df[~df['COUNTRY'].str.contains("TR")]
    df=df[~df['COUNTRY'].str.contains("XK")]
    df=df[~df['COUNTRY'].str.contains("UKC-L")]
    df=df[~df['COUNTRY'].str.contains("UKN")]
    df=df[~df['COUNTRY'].str.contains("AD")]
    df=df[~df["COUNTRY"].str.contains("BY")]
    df=df[~df["COUNTRY"].str.contains("GE")]
    nombre_paises = {
        'AL': 'Albania',
        'AT': 'Austria',
        'BA': 'Bosnia and Herzegovina',
        'BE': 'Belgium',
        'BG': 'Bulgaria',
        'CH': 'Switzerland',
        'CY': 'Cyprus',
        'CZ': 'Czech Republic',
        'DE': 'Germany',
        'DK': 'Denmark',
        'EE': 'Estonia',
        'EL': 'Greece',
        'ES': 'Spain',
        'FI': 'Finland',
        'FR': 'France',
        'HR': 'Croatia',
        'HU': 'Hungary',
        'IE': 'Ireland',
        'IS': 'Iceland',
        'IT': 'Italy',
        'LI': 'Liechtenstein',
        'LT': 'Lithuania',
        'LU': 'Luxembourg',
        'LV': 'Latvia',
        'MC': 'Monaco',
        'MD': 'Moldova',
        'ME': 'Montenegro',
        'MK': 'North Macedonia',
        'MT': 'Malta',
        'NL': 'Netherlands',
        'NO': 'Norway',
        'PL': 'Poland',
        'PT': 'Portugal',
        'RO': 'Romania',
        'RS': 'Serbia',
        'SE': 'Sweden',
        'SI': 'Slovenia',
        'SK': 'Slovakia',
        'SM': 'San Marino',
        'UA': 'Ukraine',
        'UK': 'UK'
    }
    df['COUNTRY'] = df['COUNTRY'].replace(nombre_paises)
    return df
```

```python
pib= clean_eurostat(pib)
pib.rename(columns={'VALUE': 'GDP'}, inplace=True)
pib=pib.drop(["na_item"],axis=1)
```

```python
#we will join the new df with the GDP by year and country using pandas to our original df
comb = pd.merge(df_ordered, pib, on=['COUNTRY', 'YEAR'], how='left')
comb
```

```python
#Now we have our database with our new indicator to be analyzed.
comb.to_csv("conpib.csv")
```
